```python
import pandas as pd
import numpy as np
```

```python
credits = pd.read_csv("C:\\Users\\GauravMathur\\Desktop\\Recommendation System\\tmdb_5000_credits.csv")
movies = pd.read_csv("C:\\Users\\GauravMathur\\Desktop\\Recommendation System\\tmdb_5000_movies.csv")
```

```python
credits_rename= credits.rename(index= str, columns = {'movie_id':'id'})
```

```python
movies_merge = movies.merge(credits_rename , on ='id')
movies_merge.head()
```

```python
movies_merge = movies_merge.drop(columns=['homepage','status', 'title_x','title_y','production_companies'])
```

```python
R = movies_merge['vote_average']
v = movies_merge['vote_count']
C = movies_merge['vote_average'].mean()
m = movies_merge['vote_count'].quantile(0.70)
```

```python
movies_merge['weighted_average'] = ((R*v) + (C*m)/(v+m))
```

```python
movies_merge_sorted = movies_merge.sort_values('weighted_average',ascending= False)
```

```python
movies_merge_sorted[['original_title', 'weighted_average']].head(10)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
movie_scaled = scaler.fit_transform(movies_merge[['weighted_average','popularity']])
movie_normalized = pd.DataFrame(movie_scaled,columns= ['weighted_average','popularity'])
movie_normalized.head()                                
```

```python
movies_merge[['normalized_average', 'normalized_popularity']] = movie_normalized
```

```python
movies_merge.info()
```

```python
movies_merge['score'] = (movies_merge['normalized_average']*0.5 + movies_merge['normalized_popularity']*0.5)
```

```python
movies_merge_sorted = movies_merge.sort_values('score',ascending = False)
movies_merge_sorted[['original_title', 'score']].head(20)
```
